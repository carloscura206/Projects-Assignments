# 1.5 Functions and Data transofrmatons

```python
def timeConversion(x):
    #x will be plugged into both variables seperately below
    minuteConversion = x % 100
    hourConversion = x // 100
    
    #convert minutes into hours
    fractionOfHours = minuteConversion / 60
    
    #combine the 2 variables
    totalHours = hourConversion + fractionOfHours
    
    #round to two decimal places
    roundValue = round(totalHours, 2)
    return roundValue

firstTime = timeConversion(1015)
print(firstTime)

secondTime = timeConversion(345)
print(secondTime)

```
